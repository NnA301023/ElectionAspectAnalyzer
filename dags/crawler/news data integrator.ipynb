```python
# Load libraries
import requests 
from tqdm import tqdm
from bs4 import BeautifulSoup
from newspaper import Config, Article
from urllib.parse import unquote, urlparse
```

```python
# Instantiate object
config = Config()
config.browser_user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
```

```python
# Define function to pass and retrieve news embed from RSS link
def parsing_rss_url(rss_url):
    parse_url = None
    try:
        resp = requests.get(rss_url, timeout=30)
        soup = BeautifulSoup(resp.content, "html.parser")
        links = soup.find_all("link")
        for link in links:
            link = link.get("href")
            # TODO: Implement proper logic.
            try:
                if "embed" in link:
                    parse_url = unquote(urlparse(link).query.split("=")[1])
                    break
                if "amp" in link:
                    parse_url = unquote(link)
                    if parse_url.startswith("//"):
                        parse_url = "http:" + parse_url
            except Exception as E:
                print(f"Error Occurede, Required Improve Link Parser: {links}")
    except Exception as E:
        print(f"Connection Broken with Error: {E}")
    return parse_url

def generate_content_author(news_url):
    try:
        parse_url = parsing_rss_url(news_url)
        article = Article(parse_url, config=config)
        article.download()
        article.parse()
        return article.text, ", ".join(article.authors)
    except Exception as E:
        print(f'[ERROR] - {E}')
        return np.nan, np.nan
```

### Gibran Rakabuming

```python
# Define path dataset
path_dataset = [
    "past_dataset\result_news_Gibran_Rakabuming.csv",
    "result_news_Gibran_Rakabuming.csv"
]
target_result = "final_dataset/news_gibran_rakabuming.csv"
```

```python
# Read dataset
data_1 = pd.read_csv(path_dataset[0].replace("\r", "/r"))
data_1 = data_1.replace("", np.nan)
```

```python
# Read dataset
data_2 = pd.read_csv(path_dataset[1].replace("\r", "/r"))
data_2 = data_2.replace("", np.nan)
```

```python
# Overview dataset
data_1.shape, data_2.shape
```

```python
# Overview dataset
data_1.isna().sum()
```

```python
# Overview dataset
data_2.isna().sum()
```

```python
# Overview dataset
data_1.head()
```

```python
# Overview dataset
data_2.head()
```

```python
# Concatenate dataset
data = pd.concat([data_1, data_2]).drop_duplicates(subset = ['url']).reset_index(drop = True)

# Check dimension
data.shape
```

```python
# Overview dataset
data.isna().sum()
```

```python
# Overview dataset
data.head()
```

```python
# NOTE: We can see 'content' column is NaN at all, we should fill missing value based on url given.
news_url = data['url'][0]
news_url
```

```python
# Define content and author
contents, authors = [], []
```

```python
# Enrichment for missing values content
for url in tqdm(data['url'], desc="Enrichment"):
    content, author = generate_content_author(news_url=url)
    contents.append(content)
    authors.append(author)
```

```python
# Check result
len(contents), len(authors), len(data)
```

```python
# Assign result
data['content'] = contents
data['author']  = authors
```

```python
# Overview dataset
data.isna().sum() # Great! we reduce from 1k rows into 276 rows of NaN
```

```python
# Drop NaN row on conteont and author
data_final = data.dropna(subset=['content', 'author']).reset_index(drop=True)
```

```python
# Overview dataset
data_final.isna().sum()
```

```python
# Save Result
data_final.to_csv(target_result, index=False)
```
